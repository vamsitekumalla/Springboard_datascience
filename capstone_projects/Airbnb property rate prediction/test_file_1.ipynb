### Introduction



**Problem background:**

Airbnb is an online marketplace where members can use the service to book lodging or homestays for short or long durations.

  

For this project, the chosen location is San Francisco, CA. San Francisco is the cultural, commercial and financial center of northern California. Being the 13th most populous city in the United States, San Francisco has several multinational companies headquartered in the city. The city boasts of various world renowned tourist attractions which draw travelers from all across the world.

  

**Problem Description**

Domestic and International travelers who travel for work or vacation generally tend to reside closer to the place of work/points of attraction. Several Airbnb properties are available for rent in San Francisco area with different price range and amenities.

One challenge that Airbnb hosts face is determining the optimal nightly rent price. In many areas, guests are presented with a good selection of listings and can filter on criteria like price, number of bedrooms, room type, and more. The amount a host can charge on a nightly basis is closely linked to the dynamics of the marketplace.

  

With the available data, this project is attempting to find correlation between the available amenities, location, reviews and price of the property. This Project examines and analyzes the data using tools available in Python to explore the following:

1.  Demographics of Airbnb listings  
    
2.  Exploratory analysis of property listings with the available features  
    
3.  With the above analyses, predicting the price of a new Airbnb property in the area  
    

Combining all of the above, it is attempted to build a reliable model which can predict the nightly rate of an existing or a new Airbnb property in the area using appropriate Machine Learning algorithms
  

**Data**


All of the Data is acquired from [Inside Airbnb](http://insideairbnb.com/get-the-data.html) website for San Francisco.

  

Several datasets which are available for a given city, as listed, are used :

-   Listings
    
-   Calendar
    
-   Reviews
    
-   Neighborhoods
    


-   Listings  - Contains 7151 listings in San Francisco updated as on 6 Mar ‘19 with property details including Summary, description and cancellation policy
    
-   Calendar - includes 7151 properties(listing ids) with date, availability, price, min and max nights
    
-   Neighborhood - San Francisco geojson file with marked neighborhoods with latitude, longitude data
